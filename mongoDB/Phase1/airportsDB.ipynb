```python
import pymongo
from pymongo import MongoClient
import json
from bson import json_util
# Use default client which will defaults to the MongoDB instance 
# that runs on the localhost interface on port 27017
client = MongoClient()
# db = client.sampleDB
```

```python
# Get the sampleDB database
db = client.pymongo_test
```

```python
# upload the json file
with open("airports.json","r") as f:
    data = json_util.loads(f.read())
```

```python
# Set up a new databas called "airposts"
airposts = db.airposts
result = airposts.insert_many(data)
```

```python
# Attribute of my database
print(airposts)
```

```python
# Sort data
country_sort = airposts.find().sort([
    ("country", pymongo.ASCENDING),
])
for ele in country_sort:
    print(ele["country"])
```

```python
cursor = airposts.find({"country": "China"})
for ele in cursor:
    print("Name: {}. City: {}. State: {}. ".format(ele["name"], ele["city"], ele["state"]))
```

```python
cursor = airposts.find({"country": "China", "state":"Xinjiang"})
for ele in cursor:
    print("Name: {}. City: {}. State: {}. ".format(ele["name"], ele["city"], ele["state"]))
```

```python
# We use the first airport data to perform update and remove operation.
cursor = airposts.find({"country": "China", "state":"Xinjiang"})
cursor[0]
```

```python
# Update the "state" of the first airport
update = airposts.update_one(
    {"city": "Altay"},
    {
        "$set": {
            "state": "Xinjiang (NEW!!!!!!!!)"
        },
        "$currentDate": {"lastModified": True}
    }
)
```

```python
update.matched_count
```

```python
update.modified_count
```

```python
cursor_new = airposts.find({"country": "China"})
cursor_new[0]
```

```python
# Now I want to remove this airport
airposts.count({'state': "Xinjiang (NEW!!!!!!!!)"})
```

```python
remove = airposts.delete_many({'state': "Xinjiang (NEW!!!!!!!!)"})
remove.deleted_count
```

```python
# Now we find all airports in China, there is no Altay Airport any more.
cursor = airposts.find({"country": "China"})
for ele in cursor:
    print("Name: {}. City: {}. State: {}. ".format(ele["name"], ele["city"], ele["state"]))
```
